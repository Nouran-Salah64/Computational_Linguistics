```python
# sentence tokenization ARB
import pyarabic.araby as araby
import regex as re
from nltk import sent_tokenize

file_ARB = open(r'D:\ノーラン\COLLEGE\4th year\computational 2\TASK 1 COMP\ARABIC.txt','r', encoding='utf-8')
txt_ARB = file_ARB.read()
file_ARB.close()
txt_ARB = re.sub("\n", " ", txt_ARB)

#* using rule based (regex)
arb_sent = re.split(r'(?<=[.!؟])\s?', txt_ARB) 
print("Number of sentences:", len(arb_sent), '\n')
for idx, sent in enumerate(arb_sent, start=1):
    print(f"Rule-Based Sentence {idx}: {sent}")
```

## sentece tokenization

```python
#* using pyarabic
sentences_pyarabic = araby.sentence_tokenize(txt_ARB)
print("Number of sentences:", len(sentences_pyarabic), '\n')
for idx, sent in enumerate(sentences_pyarabic, start=1):
    print(f"PyArabic Sentence {idx}: {sent}")
```

```python
# sentence tokenization ENG
import regex
import nltk
file_ENG = open(r'D:\ノーラン\COLLEGE\4th year\computational 2\TASK 1 COMP\ENGLISH.txt','r', encoding='utf-8')
txt_ENG = file_ENG.read()
file_ENG.close()
txt_ENG = re.sub("\n", " ", txt_ENG)

#* Tokenization Using regex
sentences_regex_eng = [sent.strip() for sent in regex.split(r'(?<!\b(?:Mr|Mrs)\.\s?)(?<=[.!?])\s+', txt_ENG) if sent.strip()]

print("Number of sentences:", len(sentences_regex_eng), '\n')
for idx, sent in enumerate(sentences_regex_eng, start=1):
    print(f"Rule-Based Sentence {idx}: {sent}")

#* Tokenization Using NLTK
sentences_nltk = nltk.sent_tokenize(txt_ENG)
print("Number of sentences:", len(sentences_nltk), '\n')
for Esent in sentences_nltk:
    print(Esent)
    
#* Print sentences tokenized by NLTK
for idx, sen in enumerate(sentences_nltk, start=1):
    print(f"NLTK Sentence {idx}: {sen}")
```

```python
#* Tokenization Using NLTK
sentences_nltk = nltk.sent_tokenize(txt_ENG)
print("Number of sentences:", len(sentences_nltk), '\n')
for Esent in sentences_nltk:
    print(Esent)
    
#* Print sentences tokenized by NLTK
for idx, sen in enumerate(sentences_nltk, start=1):
    print(f"NLTK Sentence {idx}: {sen}")
```
