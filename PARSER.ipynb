```python
import nltk
from nltk.tree import Tree

# Define the combined grammar
combined_grammar = nltk.CFG.fromstring("""

S ->  SV | SN | INTERROGPART SV PUNC | SN SV | SV PP |SV SN 
SN -> NP V ADVP | NP PP | EMPHATICPART NP SV | FUTUREPART NP | NP V POSS NP | NP SV | NP ADJP | NP PP | NP CONJ NP | EMPHATICPART SN | NP ADJP
SV -> V | V NP  |PREP SV| CONJ V NP PP NP | CONJ SV | V SV |V SN | V NP NP | V NP PP V NP | NP SV  CONJ V | SV PP | V NP | V PP | V PP ADVP |V NP NP|FUTUREPART V PP | CONJ V NP SV| SV CONJ NP| NEG_PARTICLE SV |NEG_PARTICLE NP SV| SV NP| CONJ V NP | CONJ V NP NP PP | CONJ V NP NP | NEGPART V PP| V SN | P SV | PROPN SV | V NP ADJP | V SV | SV PROPN 
NP -> N ADJP| N NP | DET N | N PRPERN | N N | DIRPRO DET N | PROPERN PROPERN | N CONJ N | ADJP PROPERN | N POSS | N | N DET N ADJP | PROPERN | N PROPERN | DIRPRO NP | PROPERN CONJ PROPERN CONJ PROPERN | DET N ADJP | EMPHATICPART PRONOUN | DET N ADJP | N N DET N ADJP | N PRONOUN |N NP | N POSS | N PP | N ADJP | N DET N | DET N CONJ DET N ADJP 
PP -> PREP SN | PREP NUMERAL | PREP NP NUMERAL | PREP NP | PREP SN                                       
ADJP -> ADJ | DET ADJ | ADV NP
                                       
N ->  'قلق' | 'مسألة'| 'تعليق'| 'حكومة'| 'لاعب'| 'قادمين'| 'أربعاء'| 'إياب'| 'لقاءات'| 'أسره'| 'عالم'| 'اتجاه' | 'كرواتيا' | 'بلجيكا' | 'قصراً'|'قطر'|'أمير'|'الصين'|'روسيا'|'عمل'|'تعابير' |'تداولات' | 'سوق' |'السبع' | 'مجموعة' | 'روسيا' | 'عودة' | 'الأوروبيون' | 'إصلاحات' | 'فيلم' | 'الوحش' | 'الجميلة' | 'أرباح' | 'طعام' | 'مصدر' | 'ناس' | 'معظم' | 'السعودية' | 'نجران' | 'اليمن' | 'صاروخا' | 'الحوثيين' | 'تسليح' | 'نفي' | 'إيران' | 'أمريكا' | 'نفي' | 'نظام' | 'مفاوضات' | 'وجود' | 'معارضة' | 'فصائل' | 'عام' | 'طريق' | 'مقتل' | 'شوط' | 'نهاية' | 'تقدم' | 'البرازيل' | 'سوريا' | 'أهدافا' | 'فرنسا' | 'بريطانيا'| 'أمر' | 'الغوطة' | 'مسلحو' | 'غزة' | 'قطاع' | 'منشورات' | 'إسرائيل' | 'انتخابات' | 'مقاطعة' | 'أحزاب' | 'ولد' | 'شمس' | 'لون' | 'فتاة' | 'دروس' | 'تلميذ' | 'قرد' | 'شجرة' | 'حديقة' | 'بطاقة' | 'عفرين'| 'جمهوريين' | 'انتقادات'| 'تقرير' | 'الديمقراطيون' | 'أسهم' | 'حركات' | 'جسدك' | 'ثلاثاء'
V -> 'رفضت'|'يدعم'|'تعرب'| 'تقام' | 'يزعج'|'يسحر'| 'خالفتا'| 'يهدي'|'يرغب' |'تبالغ' |'يعارضون' | 'ستنفذ' | 'أضافت' | 'يعزز' | 'يعني' | 'قالت' | 'استهدف' | 'تعترض' | 'دأبت' | 'أشار' | 'تنفي' | 'كان' | 'سيحقق' | 'قال' | 'يقترب' | 'توقف' | 'يقصفون' | 'يتجه' | 'تلقي' | 'نام' | 'يذاكر' | 'يلعب' | 'يتلقى' | 'تقرر' | 'استغل' | 'توجيه' | 'تتعافى'
DET -> 'ال'
ADJ -> 'البريطانية'| 'طائراً'|'متقلبة' | 'هراء' | 'مغلقاً' | 'أول' | 'تحذيرية' | 'مقبلة' | 'رئاسة' | 'صفراء' | 'مجتهدة' | 'محبوبة' | 'مجتهد' | 'المغربي' | 'أصفر' | 'مصرية' | 'الأمريكية'
ADV -> 'بعد' | 'مع'
DIRPRO -> 'هذا'
PREP -> 'لل' | 'منذ' | 'من' | 'إلى' | 'على' | 'في' | 'فوق' | 'مع' | 'ل' | 'بعد' | 'ب' | 'عن' 
PROPERN -> 'إسرائيل' | 'غزة' | 'الغوطة' | 'عفرين' | 'أمريكا' | 'بريطانيا' | 'فرنسا' | 'سوريا' | 'البرازيل' |'كريم الأحمدي' | 'رزان' | 'والت ديزني' | 'ترامب' | 'تميم' | 'أردوغان' | 'نيمار' | 'نايكي' | 'يوفنتوس' 
NUMERAL -> '2014' | '1958'
INTERROGPART -> 'هل'

POSS -> 'ها' |'هم'| 'ه'    
CONJ -> 'و' | 'أو'
NEG_PARTICLE -> 'لا' | 'لكن'
EMPHATICPART -> 'إن' | 'أن'
PRONOUN -> 'ه'
FUTUREPART -> 'س'
PUNC -> '.' | '؟'         
""")

EXCEPTION_WORDS = [
    'المغربي', 
    'الأحمدي',
    'البرازيل',
    'الحوثيين',
    'السعودية',
    'الغوطة',
    'اليمن',
    'الجميلة',
    'الوحش',
    'الأوروبيون',
    'السبع',
    'الديمقراطيون',
    'الأمريكية',
    'لاعب',
    'الصين',
    'لكن',
    'لا',
    'البريطانية'
]

def arabic_tokenizer(sentence, grammar):
    if sentence in EXCEPTION_WORDS:
        return [sentence]
    
    tokens = []
    while sentence:
        matched = False
        
        for word in EXCEPTION_WORDS:
            if sentence.startswith(word):
                tokens.append(word)
                sentence = sentence[len(word):].strip()
                matched = True
                break
        
        if matched:
            continue
        
        for production in grammar.productions():
            if production.is_lexical():
                word = production.rhs()[0]
                if sentence.startswith(word):
                    tokens.append(word)
                    sentence = sentence[len(word):].strip()
                    matched = True
                    break
                
        if not matched:
            raise ValueError(f"Unable to tokenize: {sentence}")
    
    return tokens

# Example usage
sents = [
    "الولد مجتهد",
    "الشمس لونها أصفر",
    "الفتاة المجتهدة محبوبة",
    "نام الولد",
    "التلميذ يذاكر دروسه",
    "القرد فوق الشجرة",
    "الولد في الحديقة",
    "الولد يلعب في الحديقة",
    "المغربي كريم الأحمدي يتلقى بطاقة صفراء",
    "سوق الأسهم الأمريكية تتعافى بعد تداولات متقلبة",
    "لا تبالغ في حركات جسدك أو التعابير",
    'أمير قطر تميم يهدي أردوغان قصراً طائراً',
    "ترامب يرغب في ال عمل مع روسيا والصين",
    "لكن بلجيكا وكرواتيا خالفتا هذا الاتجاه",
    "نيمار يسحر البرازيل ويزعج العالم بأسره",
    "وتقام لقاءات الإياب الثلاثاء والأربعاء القادمين",
    "نايكي تعرب عن قلقها ويوفنتوس يدعم اللاعب",
    "ورفضت الحكومة البريطانية التعليق على المسألة",
    'أحزاب مصرية تقرر مقاطعة انتخابات الرئاسة المقبلة',
    'إسرائيل تلقي منشورات تحذيرية على قطاع غزة',
    'هل يتجه مسلحو الغوطة إلى عفرين ؟',
    'أمريكا وبريطانيا وفرنسا يقصفون أهدافا في سوريا',
    'البرازيل توقف تقدم فرنسا في 1958',
    'الشوط الأول يقترب من النهاية',
    'وقال إنه سيحقق في مقتل رزان',
     'وكان الطريق مغلقاً منذ عام 2014'
    "فصائل المعارضة تنفي وجود مفاوضات مع النظام",
    "وأشار إلى أن نفي الأمر هراء",
    "ودأبت إيران على نفي تسليح الحوثيين",
    "السعودية تعترض صاروخا من اليمن استهدف نجران",
    "وقالت معظم الناس يعنيهم مصدر طعامهم",
    "فيلم الجميلة والوحش يعزز أرباح والت ديزني",
    "وأضافت أن الإصلاحات ستنفذ هذا العام",
    "الأوروبيون يعارضون عودة روسيا إلى مجموعة السبع",
    "واستغل الديمقراطيون التقرير لتوجيه الانتقادات للجمهوريين",

]

# Create a parser object
parser = nltk.ChartParser(combined_grammar)

for sent in sents:
    try:
        tokens = arabic_tokenizer(sent, combined_grammar)
        print(f"Tokens: {tokens}\n")
        trees = parser.parse_all(tokens)
        for tree in trees:
            print(tree)
            tree.draw()
    except ValueError as e:
        print(f"Error: {e}")
```
