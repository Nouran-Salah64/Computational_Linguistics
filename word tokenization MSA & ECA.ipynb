```python
# sentence tokenization MSA
import regex as re
file_ARB = open(r'C:\Users\test\Desktop\TASK 2 COMP\2000 MSA.txt','r', encoding='utf-8')
txt_ARB = file_ARB.read()
file_ARB.close()
txt_ARB = re.sub("\n", " ", txt_ARB)

sentences = re.split(r'[.؟!]',txt_ARB)
print("Number of sentences:", len(sentences), '\n')
print(sentences)
```

```python
# word tokenization MSA
text = " ".join(sentences)
word_tokenization = re.findall(r'[.؟!؛:"،]|[ء-ي]+',text)
print("Number of tokens:", len(word_tokenization), '\n')
print(word_tokenization)
```

```python
# sentence tokenization ECA
import regex as re
file_arb=open(r'C:\Users\test\Desktop\TASK 2 COMP\2000 ECA.txt','r',encoding='utf-8')
text_arb=file_arb.read()
file_arb.close()
text_arb = re.sub("\n", " ", text_arb)
arb_sent = re.split(r'(?<=[.؟!])\s?',text_arb)
print("Number of sentences:", len(arb_sent), '\n')
print(arb_sent)
```

```python
# word tokenization ECA
text = " ".join(arb_sent)
word_tokenization = re.findall(r'[.؟!؛:"،]|[ء-ي]+',text)
print("Number of tokens:", len(word_tokenization), '\n')
print(word_tokenization)
```
